```python
import numpy as np
import statsmodels.api as sm
from scipy import stats
from matplotlib import pyplot as plt
from sklearn import neighbors
from   sklearn.metrics import r2_score
import pylab as pl

import os
import psycopg2
import pandas as pd
from fetch_station import fetch_station

# Fetch all records for a single station (tfl_id == 17 in DC)
bike_data_bucketed = fetch_station('Washington, D.C.',17,15, 'max')
```

```python
print bike_data_bucketed.head()

# calculate percent full

#first convert to arrays so that Python doesn't do integer division
bikes_available=np.asarray(bike_data_bucketed["bikes_available"], dtype=np.float32)
slots_available=np.asarray(bike_data_bucketed["slots_available"], dtype=np.float32)

pct_full = bikes_available / (bikes_available + slots_available)

print "number of total values " + str(len(pct_full))

pct_full = pct_full[np.isfinite(pct_full)]

print "number of values after infinite removed " + str(len(pct_full))
```

```python
# function for setting up time lagged variables

def change_mat(ts,d):
    """Recieve a time series and a markov delay period.
       Return a matix of that design."""
    n=len(ts)
    x = np.array(ts[(d-1):n-1])
    for r in range(2,d+1):
        b = np.array(ts[(d-r):(n-r)])
        x = np.concatenate([x.reshape(len(x),-1),b.reshape(len(b),-1)],axis=1)
    y = ts[d:]
    return x, y
```

```python
x, y = change_mat(pct_full, 8)

num_train = 4*len(pct_full) // 5
#num_train = 100
num_test  = 4 * 24 * 7

x_train = x[:num_train]
y_train = y[:num_train]
x_test  = x[num_train : num_train + num_test]
y_test  = y[num_train : num_train + num_test]
```

```python
n_neighbors = 5

knn = neighbors.KNeighborsRegressor(n_neighbors, weights='uniform')
y_ = knn.fit(x_train, y_train).predict(x_test)

preds = knn.predict(x_test)
knn.get_params()

plt.plot(range(num_test), y_test, label = "Truth")
plt.plot(range(num_test), preds,  label = "Preds")
plt.legend()
r2 = r2_score(y_test, preds)
plt.title("R2 score: %.2f" %r2)
plt.show()
```
